```python
#dice(2times) = [1,2,3,4,5,6]
print("Omega: ")
#P > 10 --> 11,12
def omega(lista):
  A=[]
  for i in range(1,7):
    for y in lista:
      A.append((y,i))
    print(A,end=' ')
    print()

  B = [(y,i) for (y,i) in A if (y+i)>10]
  print(B)
  print('B', len(B))
  print('A', len(A))
  P = (len(B)/len(A))
  print(f"Probabilty of getting sum grater than 10 is: {P:.2f}")
omega((1,2,3,4,5,6))
```

```python
#dice(2times) = [1,2,3,4,5,6]
print("Omega: ")
#P(sum^2) >= 45 
dice = (1,2,3,4,5,6)
A=[]
for i in range(1,7):
  for y in dice:
    A.append((y**2,i**2))
  print(A,end=' ')
  print()

B = [(y,i) for (y,i) in A if (y+i)>=45]
print(B)
print('B', len(B))
print('A', len(A))
P = (len(B)/len(A))
print(f"Probabilty of getting sum grater than 45 is: {P:.2f}")
```

```python
dice = (1,2,3,4,5,6)
A=[]
for i in range(1,7):
    for j in range(1,7):
      for y in dice:
        A.append((y,j,i))

print(A,end=' ')
print()
print(len(A))
B = [(y,j,i) for (y,j,i) in A if (y+j+i)%7==0]
print(B)
print(len(B))
P = (len(B)/len(A))
print(f"Probabilty of getting sum mod7 = 0 is: {P:.2f}")
```

```python
dice = (1,2,3,4,5,6)
A=[]
for i in range(1,7):
    for j in range(1,7):
      for y in dice:
        A.append((y,j,i))
print(A,end=' ')
print()
print(len(A))
B = [(y,j,i) for (y,j,i) in A if (y**2+j**2+i**2)%3==0]
print(B)
print(len(B))
P = (len(B)/len(A))
print(f"Probabilty sum mod3 is: {P:.4f}")
```
